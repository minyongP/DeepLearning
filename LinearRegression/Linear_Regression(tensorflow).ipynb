- 선형회귀

```python
import numpy as np
import matplotlib.pyplot as plt

from keras.models import Sequential
from keras.layers import Dense
```

```python
x = np.array([2, 4, 6, 8])      # 공부 시간
y = np.array([81, 93, 91, 97])  # 시험 점수

model = Sequential()

model.add(Dense(1, input_dim=1, activation='linear'))

#오차 수정(그레디언트):sdg, 오차 정도: mse
model.compile(optimizer='sgd', loss='mse')

model.fit(x, y, epochs=2000)

plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r')
plt.show()

#임의의 시간을 넣어 점수를 예측하는 모델
hour=7  

prediction = model.predict([hour])
print("%.f 시간을 공부할 경우의 예상 점수는 %.02f점입니다." % (hour, prediction))
```

- 다중 선형 회귀

```python
x = np.array([[2, 0], [4, 4], [6, 2], [8, 3]])      # 공부 시간, 과외 시간
y = np.array([81, 93, 91, 97])  # 시험 점수

model = Sequential()

model.add(Dense(1, input_dim=2, activation='linear'))
model.compile(optimizer='sgd', loss='mse')

model.fit(x, y, epochs=2000)

#임의의 학습시간과 과외 시간을 입력해 점수를 예측하는 모델
hour = 7  
private_class = 4
prediction = model.predict([[hour, private_class]])
print("%.f 시간을 공부하고 %.f 시간의 과외를 받은 경우의 예상 점수는 %.02f점입니다." % (hour, private_class, prediction))
```
