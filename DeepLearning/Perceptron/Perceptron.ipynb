퍼셉트론이 개발되고 아달라인이 보완하며 인공지능 발전 시작. 하지만 XOR문제를 해결하지 못하는 문제가 있음

해결책은 다층 퍼셉트론과 오차 역전파

- 다층 퍼셉트론

> 퍼셉트론 두개를 한번에 계산(hidden layer)



```python
import numpy as np
```

```python
# 다층 퍼셉트론으론 XOR문제 해결
# NAND와 OR를 hidden layer에서 동시에 계산한 후 AND로 출력해주면 된다.

# 가중치와 바이어스
w11 = np.array([-2, -2])
w12 = np.array([2, 2])
w2 = np.array ([1, 1])

b1 = 3
b2 = -1
b3 = -1

# 퍼셉트론
def MLP(x, w, b):
  y = np.sum(w * x) + b
  if y <= 0:
    return 0
  else:
    return 1

# NAND 게이트
def NAND(x1, x2):
  return MLP(np.array([x1, x2]), w11, b1)

# OR 게이트
def OR(x1, x2):
  return MLP(np.array([x1, x2]), w12, b2)

# AND 게이트
def AND(x1, x2):
  return MLP(np.array([x1, x2]), w2, b3)

# XOR 게이트
def XOR(x1, x2):
  return AND(NAND(x1, x2), OR(x1, x2))

# x1, x2 값을 번갈아 대입하며 최종 값 출력
for x in [(0, 0), (1, 0), (0, 1), (1, 1)]:
  y = XOR(x[0], x[1])
  print("입력 값: " + str(x) + "출력 값: " + str(y))
```

- 오차역전파(backpropagation)

> 다층 퍼셉트론으로는 은닉층에 포함된 가중치를 업데이트할 방법이 없었다. 하지만 오차역전파를 이용하여 가중치를 업데이트 가능하게 계산법이 생겼다.



- 렐루(Lelu)

> 기존의 시그모이드 함수는 깊은 층을 만들수록 미분된 값이 0에 가까워진다. 이를 해결하기 깊은 층을 만들어도 사용 가능한 활성함수




- 아담(Adam)
