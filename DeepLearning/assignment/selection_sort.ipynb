```python
def selection_sort(arr):              #선택정렬
    for i in range(len(arr)):
        sel = i
        for j in range(i+1,len(arr)):
            if arr[i]<arr[j]: sel = j
        temp = arr[sel]
        arr[sel] = arr[i]
        arr[i] = temp
    return arr

num_list = []
new_num = int(input("자연수 한 개를입력하세요." ))
num_list.append(new_num)

i=0
while(True):
  i += 1

  if i == 100:
   print("이제그만하세요. 벌써 100개를 입력했어요.")
   break
  
  new_num = int(input("계속입력을 원하시면, 자연수 한 개를입력하시고, 아니라면 0을입력하세요." ))

  if new_num == 0:
    break
  
  num_list.append(new_num)

num_list = selection_sort(num_list)
print('당신이 입력한 숫자는 큰 것부터 순서대로 말한다면 아래와 같습니다.')

for i in num_list:
  print(i, end =', ')
```
