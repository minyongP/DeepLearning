# 로지스틱 회귀 모델
- 결과값이 0과 1로만 출력됨
  0과 1로만 값이 나와 직선으로 표현 불가능함.
- sigmoid 함수를 이용해 표현 [y=1/(1+e^(-(ax+b)]
- sigmoid의 a,b값을 구하기 위해서는 mse가 아닌 로그함수를 이용해야함 그렇기 위해 '교차 엔트로피 함수'를 이용

```python
import numpy as np
import matplotlib.pyplot as plt

from keras.models import Sequential
from keras.layers import Dense
```

```python
x = np.array([2, 4, 6 ,8, 10, 12, 14])
y = np.array([0, 0, 0, 1, 1, 1, 1])

model = Sequential()
model.add(Dense(1, input_dim=1, activation='sigmoid'))

#교차 엔트로피 오차 함수를 이용하기 위해 'binary_crossentropy'로 설정
model.compile(optimizer='sgd', loss='binary_crossentropy')
model.fit(x, y, epochs=5000)
```

```python
plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r')
plt.show()
```

학습이 진행됨에 따라 그래프가 점점 시그모이드 함수 그래프의 형태를 취한다.

```python
hour = 7
prediction = model.predict([hour])
print("%.f 시간을 공부할 경우의 합격 예상 확률은 %.02f점입니다." % (hour, prediction * 100))
```

학습시간이 7시간보다 클 경우에는 합격 확률이 50%가 넘는 것을 알 수 있다.
(데이터 양, 학습 시간 등 환경에 따라 예측 정확도는 향상 가능하다.)
