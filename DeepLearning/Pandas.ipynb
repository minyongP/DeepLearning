```python
# 새로운 열을 추가하기 데이터 프레임
df = pd.DataFrame(
    {'a':[4,5,6,7],
     'b':[8,9,10,11],
     'c':[12,13,14,15]},
     index=pd.MultiIndex.from_tuples(
         [('d',1),('d',2),('e',1),('e',2)],
         names=['n','v']
     )
)

df
```

```python
# 조건에 맞는 열 추가
df['sum'] = df['a'] + df['b'] + df['c']

df
```

```python
# assign()을 이용해 조건에 맞는 열 추가
df = df.assign(multiply=lambda df: df['a']*df['b']*df['c'])

df
```

```python
# 데이터 병합 데이터 프레임
adf = pd.DataFrame({"x1" : ["A","B","C"], 
                    "x2": [1,2,3]})
adf
```

```python
bdf = pd.DataFrame({"x1" : ["A","B","D"], 
                    "x3": ["T","F","T"]})
bdf
```

```python
cdf = pd.DataFrame({"x1" : ["B","C","D"], 
                    "x2": [2,3,4]})
cdf
```

```python
# 왼쪽 열을 축으로 병합
pd.merge(adf, bdf, how='left', on='x1')
# x1을 키로 해서 병합, 왼쪽(adf)를 기준으로
# 왼쪽의 adf에는 D가 없으므로 해당 값은 NaN으로 변환 
```

```python
# 오른쪽 열을 축으로 병합
pd.merge(adf, bdf, how='right', on='x1')
# x1을 키로 해서 병합, 오른쪽(bdf)을 기준으로
# 오른쪽의 bdf에는 C가 없으므로 해당 값은 NaN으로 변환 
```

```python
# 공통 값만 병합
pd.merge(adf, bdf, how='inner', on='x1')
```

```python
# 모든 값을 병합
pd.merge(adf, bdf, how='outer', on='x1')
```

```python
# 특정한 열을 비교해서 공통 값이 존재하는 경우만 가져오기
adf[adf.x1.isin(bdf.x1)]
# adf와 bdf의 특정한 열을 비교하여 공통 값이 존재하는 경우만 가져오기
# adf.x1열과 bdf.x1열은 A,B가 같다. 따라서 adf의 해당 값만 출력
```

```python
# 공통 값이 존재하는 경우 해당 값을 제외하고 병합
adf[~adf.x1.isin(bdf.x1)] # 해당 값만 제외하고 출력
```

```python
# 공통 값이 있는 것만 병합
pd.merge(adf, cdf)
```

```python
# 모두 병합
pd.merge(adf, cdf, how='outer')
```

```python
# 어디서 병합되었는지 표시
pd.merge(adf, cdf, how='outer', indicator=True)
```

```python
# 원하는 병합만 남기기
pd.merge(adf, cdf, how='outer', indicator=True).query('_merge == "left_only"')
```

```python
# merge 칼럼 제거
pd.merge(adf, cdf, how='outer', indicator=True).query('_merge == "left_only"').drop(columns=['_merge'])
```

```python
# 그룹별로 집계 데이터 프레임

# 주택 가격 예측 데이터
df = pd.read_csv("https://raw.githubusercontent.com/taehojo/deeplearning/master/data/house_train.csv")
df
```

```python
# 그룹 지정 및 그룹별 데이터 수 표시
df.groupby(by='YrSold').size()  # 연도 중심으로 그룹별 수 표시
```

```python
# 그룹 지정 후 원하는 칼럼 표시
# 팔린 연도를 중심으로 그룹을 만든 후 각 그룹별로 주차장의 넓이 표시
df.groupby(by='YrSold')['LotArea'].mean()
```

```python
# 밀집도 기준으로 순위 부여
df['SalePrice'].rank(method='dense')
```

```python
# 최저 값을 기준으로 순위 부여
df['SalePrice'].rank(method='min')    
```

```python
# 순위를 비율로 표시하기  
df['SalePrice'].rank(pct=True) # 1에 가까울수록 비싼 집
```

```python
# 동일 순위에 대한 처리 방법 정하기 
df['SalePrice'].rank(method='first')
```
