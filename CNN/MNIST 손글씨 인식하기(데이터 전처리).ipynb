MNIST 손글씨 인식하기(데이터 전처리)

```python
from keras.datasets import mnist
from keras.utils import to_categorical
```

```python
import matplotlib.pyplot as plt
import sys
```

```python
# mnist 데이터셋을 불러와 학습셋과 테스트셋으로 저장
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
print("학습셋 이미지수: %d개" %(X_train.shape[0]))
print("테스트셋 이미지수: %d개" %(X_test.shape[0]))
```

```python
plt.imshow(X_train[0], cmap='Greys')
plt.show()
```

```python
# 컴퓨터는 어떻게 인식할까?
for x in X_train[0]:
  for i in x:
    sys.stdout.write("%-3s" % i)
  sys.stdout.write('\n')
```

```python
# 차원변환  (28x28의 2차원을 784개의 1차원 배열로 변환)
X_train = X_train.reshape(X_train.shape[0], 784)

# 정규화[normalization]  (케라스는 0과 1 사이의 값으로 구동할 때 최적의 성능을 내므로 변환)
X_train = X_train.astype('float64')   # 255로 나누기 위해 실수로 변환
X_train = X_train / 255

# 테스트 값도 마찬가지
X_test = X_test.reshape(X_test.shape[0], 784).astype('float64') / 255

# 클래스 값
print("class : %d" % (y_train[0]))

# 바이너리화 과정(0-9의 정수형 값을 갖는 현재 클래스 값에서 0과 1로만 이루어진 벡터로 수정[원-핫 인코딩])
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

print(y_train[0])
```
