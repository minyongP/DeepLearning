```python
while True:
  temp = input("30이하의 정수를 입력. : ")

  if not temp.isnumeric():                    #isdecimal: 문자열이 정수형으로 변환 가능한지 체크 / isnumeric() : 제곱, 분수까지 체크
    print("정수로 입력하세요")
    continue

  elif int(temp) > 30:
    print("너무 많다")
    continue

  else:
    break

temp = int(temp)

for a in range(1,temp+1):
  for b in range(1,a+1):
    print(b, end = " ")
  print()
```
